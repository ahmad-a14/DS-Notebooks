In `data.csv`, there are some missing values, which are denoted by `?`. List down all the columns with number of missing values as shown below

<img src="Fig2.PNG" />

```python
# Your code here (marks 4)
for col in data.columns:
    print('Column: {} Missing Values : {}'.format(col,data[data[col]=='?'].count()[0]))
```

Interesting column `Job` and column `Jobclass` contains many missing values. How many records have both these columns missing?

```python
# Your code here (marks 2)
fil=data[data['Job']=='?']
fil=fil[fil['Jobclass']=='?']
print('answer is : ',str(fil.shape[0]))
```

Replace all `?` occurrences with `NaN`

```python
# Your code here (marks 2)
data.replace('?',np.nan,inplace=True)
data
```

### Checkpoint1

List unique enteries in columns `Job`, `Jobclass` and `native-country`

```python
# Your code here (marks 3)
print('Unique entries in Job column\n')
for i in data['Job'].unique():
        print(i)
print('\nUnique entries in Jobclass column\n')
for i in data['Jobclass'].unique():
        print(i)
print('\nUnique entries in native-country column\n')
for i in data['native-country'].unique():
        print(i)
```

Replace `NaN` in columns `Job`, `Jobclass` and `native-country` with the most frequently occurred value.

Hint: Statistical function `Mode` can help in this

```python
# Your code here (marks 6)
for x in ['Job', 'Jobclass','native-country']:
    data[x]=data[x].fillna(data[x].mode()[0])
data
```

### Checkpoint2

`income` column contains strings `<=50K` and `>50K`. Replace with `0` and `1` respectively.

```python
# Your code here (marks 2)
data['income']=data['income'].replace('<=50K',0).replace('>50K','1')
data['income']
```

Convert column `income` data type to `int32`

```python
#Your code here (marks 1)
data['income']=data['income'].astype('int32')
data['income'].dtype
```
